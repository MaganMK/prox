```python
import os, pickle

path = "../data/train"
file_names = os.listdir(path)
steps = []
lim = 300
for i, file_name in enumerate(file_names):
    file_path = f"{path}/{file_name}"
    step = pickle.load(open(file_path, "rb"))
    steps.append(step)
    if i > lim:
        break
```

```python
for step in steps:
    tactic_application = step["tactic"]["text"]
    gc = []
    lc = []
    for g in step["env"]:
        gc.append(g["qualid"])
    for l in step["local_context"]:
        lc.append(l["ident"])
        
    print(gc)
    print(lc)
    print(tactic_application)
```
